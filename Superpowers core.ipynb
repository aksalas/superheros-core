```python
superpowers['powers_split'] = superpowers['Powers'].str.split(',')
superpowers['powers_split']
```

```python
exploded = superpowers.explode('powers_split')
exploded
```

```python
cols_to_make = exploded['powers_split'].dropna().unique()
cols_to_make
```

```python
for col in cols_to_make:
    superpowers[col] = superpowers['Powers'].str.contains(col)
superpowers.head()
```

```python
left_df = superpowers
right_df = superhero_info
```

```python
merged_df = pd.merge(left_df, right_df, left_on='hero_names', right_on='Hero', how='right')

merged_df.head()
```

```python
merged_df['Weight'] = merged_df['Weight'].str.replace(' kg', '')
merged_df['Height'] = merged_df['Height'].str.replace(' cm', '')
merged_df['Weight']
```

```python
merged_df['Weight'] = pd.to_numeric(merged_df['Weight'])
merged_df['Height'] = pd.to_numeric(merged_df['Height'])
```
